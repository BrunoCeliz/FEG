# 6)
#### Queremos la evolución temporal del radio de una cáscara esférica (de radio $r$ que encierra una masa $M(<r)$) en el modelo de colapso esférico. Para ello, comencemos escribiendo la ecuación de movimiento de una cáscara esférica:
### $\frac{d^2 r}{dt^2} = -\frac{GM}{r^2}$
#### Multiplicando ambos lados por $\dot{r}$ *(donde "$^{\cdot}$" denota la derivada resepcto al tiempo ($\frac{d}{dt}$))*:
### $\dot{r} \cdot \ddot{r} = -GM \frac{\dot{r}}{r^2}$
### $\dot{r} \cdot \ddot{r} + GM \frac{\dot{r}}{r^2} = 0$
#### Considerando
### $\frac{1}{2} \frac{d}{dt} \dot{r}^2 = \dot{r} \cdot \ddot{r}$; $\frac{d}{dt} \frac{-1}{r} = \frac{\dot{r}}{r^2}$
#### Podemos integrar respecto al tiempo la ecuación y nos queda
### $\frac{1}{2} (\frac{dr}{dt})^2 - \frac{GM}{r} = cte = E$; donde $E$ es la energía de la cáscara
#### Entonces, como queremos obtener $r(t)$, separemos lso diferenciales:
### $dr = \sqrt{2(E + \frac{GM}{r})} dt$
### $r(t) = \sqrt{2|E|} \int \sqrt{1 + \frac{GM}{r|E|}} dt$
#### Que es la ecuación de una curva braquistócrona (cicloide), cuya solución viene dada por la forma paramétrica
### $r(\theta) = \frac{GM}{2|E|} (1-cos(\theta))$
### $t(\theta) = \frac{GM}{(2|E|)^{3/2}} (\theta-sen(\theta))$
### $\theta \in [0,2\pi]$

# 7)

#### Sabemos que $t_{ta} = \pi B$, con $B = \frac{GM}{(2|E|)^{3/2}}$. Donde
### $E = cte = E_i = K_i + W_i$,
### $K_i = \frac{v_i^2}{2}$, donde suponemos que $v_i \approx \dot{a} x_i = H_i r_i$ *("Hubble flow")*,
### $W_i = -\frac{GM}{r_i}$ y escribimos $M(r<r_i) = cte = \frac{4}{3} \pi r_i^3 \rho_i$, con $\rho_i = \overline{\rho}_i (1+\delta_i)$; $\overline{\rho}_i = \frac{1}{6 \pi G} (\frac{3}{2} H_i)^2$
### $\Rightarrow M_i = \frac{H_i^2 r_i^3}{2G} (1+\delta_i)$ y $E_i = -K_i \delta_i = -2H_i^2 r_i^2 \delta_i$.

#### Veamos cuál es el $t_{ta}$ para 2 cáscaras diferentes provenientes de la misma sobredensidad en el modelo de colapso esférico. Esto es, cáscaras $1$ y $2$: $r_{i,1} < r_{i,2}$ y $\delta_{i,1} = \delta_{i,2}$. Luego,

### $t_{ta} = \frac{\pi GM}{(2|E|)^{3/2}}$
### $M_1 = \frac{H_i^2 r_1^3}{2G} (1+\delta_i)$; $M_2 = \frac{H_i^2 r_2^3}{2G} (1+\delta_i)$ 
### $E_1 = -2H_i^2 r_1^2 \delta_i$; $E_2 = -2H_i^2 r_2^2 \delta_i$
#### Entonces,
### $t_{ta,1} = \frac{\frac{\pi G}{2G} H_i^2 r_1^3 (1+\delta_i)}{(4H_i^2 r_1^2 \delta_i)^{3/2}}$
### $t_{ta,1} = \frac{\pi (1+\delta_i)}{16 \delta_i^{3/2} H_i}$
#### Análogamente
### $t_{ta,2} = \frac{\frac{\pi G}{2G} H_i^2 r_2^3 (1+\delta_i)}{(4H_i^2 r_2^2 \delta_i)^{3/2}}$
### $t_{ta,2} = \frac{\pi (1+\delta_i)}{16 \delta_i^{3/2} H_i}$
## $\therefore t_{ta,1} = t_{ta,2}$
#### Como $t_{colapso} = 2 t_{ta}$, también vale que $t_{col,1} = t_{col,2}$. Esto vale siempre que $0 < r_i < R$, donde $R$ es hasta donde llega la sobredensidad $\delta_i$ *Top-hat*.

#### Queremos escribir las ecuaciones de movimiento para un fluido colisional (continuidad, Euler, adiabática y masa) bajo auto-similaridad en un modelo *Secondary Infall*:
### i) $\frac{\partial \rho}{\partial t} \frac{1}{r^2} \frac{\partial}{\partial r} (r^2 \rho v) = 0$
### ii) $\frac{\partial v}{\partial t} + v \frac{\partial v}{\partial r} = - \frac{1}{\rho} \frac{\partial P}{\partial r} - \frac{GM(r,t)}{r^2}$
### iii) $(\frac{\partial}{\partial t} + v \frac{\partial}{\partial r}) ln(\frac{P}{\rho^{\gamma}}) = 0$
### iv) $\frac{\partial M}{\partial r} = 4 \pi r^2 \rho$
#### Entonces, suponemos un universo EdS, cuya perturbación inicial sigue una power-law: $\delta_i \propto M_i^{- \epsilon} \propto r_i^{-3 \epsilon}$. Bertschinger 1985 propone $\epsilon = 1$, y para auto-similaridad define $\lambda = \frac{r}{r_{ta}}$, donde $r_{ta} \propto t^{8/9}$. Luego, en su verisón auto-similar tenemos:
### $\rho(r,t) = \overline{\rho}(t) D(\lambda)$
### $P(r,t) = \overline{\rho}(t) (\frac{r_{ta}}{t})^2 P(\lambda)$
### $v(r,t) = \frac{r_{ta}}{t} V(\lambda)$
### $M(r,t) = \frac{4 \pi}{3} \overline{\rho}(t) r_{ta}^3 M(\lambda)$
#### Por lo tanto, debemos reemplazar estas expresiones en las ecuaciones anteriores y obtener un sistema de ecauciones diferenciales para $P(\lambda), D(\lambda), M(\lambda)$ y $V(\lambda)$ (variables adimensionales). Para ello, como cálculo auxiliar dejamos las consideraciones más importantes:
### $r = \lambda \cdot r_{ta}$
### $\frac{\partial }{\partial r} = \frac{\partial }{\partial \lambda} \frac{\partial \lambda}{\partial r} = \frac{\partial }{\partial \lambda} \frac{1}{r_{ta}}$
### $\frac{\partial }{\partial t} = \frac{\partial }{\partial \lambda} \frac{\partial \lambda}{\partial r_{ta}} \frac{\partial r_{ta}}{\partial t} = - \frac{8 \lambda}{9 t} \frac{\partial }{\partial \lambda}$
### $\frac{\partial \overline{\rho}}{\partial t} = -2 \frac{\overline{\rho}}{t}$
### $\frac{\partial }{\partial \lambda} (f(\lambda) \cdot g(\lambda)) = f'g + fg'$; donde "$'$" es la derivada con respecto a $\lambda$.
#### Al reemplazar y llevar a cabo las cuentas correspondientes, llegamos a:
### iv) $M' = 3 \lambda^2 D$
### i) $D' (V - \frac{8}{9} \lambda) + DV' + \frac{2DV}{\lambda} - 2D = 0$
### ii) $V'(V - \frac{8}{9} \lambda) - \frac{V}{9} = -\frac{P'}{D} - \frac{2M}{9 \lambda^2}$
### iii) $(\frac{P'}{P} - \gamma \frac{D'}{D})(V - \frac{8}{9} \lambda) = \frac{20}{9} - 2\gamma$
#### Acá *$D,P,M$ y $V$ son funciones de $\lambda$ (variables adimensionales).*
